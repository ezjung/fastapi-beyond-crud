```python
import requests

BASE_URL = "http://127.0.0.1:8000"
```

```python
def test_book_list():
    print("=== GET /books ===")
    resp = requests.get(f"{BASE_URL}/books")
    print(f"Status: {resp.status_code}")
    print(f"Body:   {resp.json()}")
    print()
```

```python
test_book_list()
```

```python
def test_read_a_book(id:int):
    print(f"=== GET /book/{id} ===")
    resp = requests.get(f"{BASE_URL}/book/{id}")
    print(f"Status: {resp.status_code}")
    print(f"Body:   {resp.json()}")
    print()
    
```

```python
test_read_a_book(1)
```

```python
test_read_a_book(100)
```

```python
from pydantic import BaseModel

class Book(BaseModel):
    id: int
    title: str
    author: str
    publisher: str
    published_date: str
    page_count: int
    language: str
```

```python
def create_a_book(book_data: Book):
    print(f"=== POST /books ===")
    resp = requests.post(f"{BASE_URL}/books", json=book_data.model_dump())
    print(f"Status: {resp.status_code}")
    print(f"Body:   {resp.json()}")
    print()

```

```python
create_a_book(Book(id=100, title="The Great Gatsby", author="F. Scott Fitzgerald", publisher="Scribner", published_date="1925-04-10", page_count=180, language="English"))
```

```python
class BookUpdateSchema(BaseModel):
    title:str
    author:str
    publisher:str
    page_count:int
    language:str
```

```python
def update_a_book(id:int, book_update_data:BookUpdateSchema):
    print(f"=== PATCH /book/{id} ===")
    resp = requests.patch(f"{BASE_URL}/book/{id}", json=book_update_data.model_dump())
    print(f"Status: {resp.status_code}")
    print(f"Body:   {resp.json()}")
    print()
    
```

```python
update_a_book(100, BookUpdateSchema(title="The Great Gatsby", author="F. Scott Fitzgerald", publisher="Scribner", page_count=200, language="French"))
```

```python
def delete_a_book(id:int):
    print(f"=== DELETE /book/{id} ===")
    resp = requests.delete(f"{BASE_URL}/book/{id}")
    print(f"Status: {resp.status_code}")
    print()
    
```

```python
delete_a_book(100)
```

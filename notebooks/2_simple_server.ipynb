```python
import requests

BASE_URL = "http://127.0.0.1:8000"
```

```python
def test_create_user(username: str, email: str):
    print(f"=== POST /create_user ===")
    payload = {"username": username, "email": email}
    print(f"Sending: {payload}")
    # json= parameter serialise python dict to json string; sets the content-type to application/json
    # res = requests.post(f"{BASE_URL}/create_user", payload) # Cause 422 upprocessable Entity error
    res = requests.post(f"{BASE_URL}/create_user", json=payload) # dict to json string
    print(f"Status:  {res.status_code}")
    print(f"Body:    {res.json()}")
    print()
```

```python
test_create_user('Sungsoon', 'sungsoon@example.com')
```
